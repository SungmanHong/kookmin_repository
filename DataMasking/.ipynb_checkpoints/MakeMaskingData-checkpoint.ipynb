```python
import os
from os import listdir
from os.path import isfile, join
import pandas as pd
from pandas import DataFrame as df
```

```python
block = ('A11-1','A1-4','호평','A17','A5-1','A12-1','A7','A8-1', 'A9')
```

```python
files = [f for f in listdir(os.curdir) if isfile(join(os.curdir, f))]
```

```python
type(files)
```

```python
b = "1_dang.xlsx"
if 'dang' in b:
    print("111")
```

```python
excel_files = [s for s in files if ".xlsx" in s]
```

```python
dang_list = []
shin_list = []
```

```python
for file in excel_files :
    if 'dang' in file:
        dang_list.append(file)
    else :
        shin_list.append(file)
```

```python
dang_list
```

```python
dang_frames = []
shin_frames = []
```

```python
for dang in dang_list:
    dang_excel = pd.read_excel(dang)
    dang_excel['블록'] = block[int(dang.split('_')[0])-1]
    dang_frames.append(dang_excel)

for shin in shin_list:
    shin_excel = pd.read_excel(shin)
    shin_excel['블록'] = block[int(shin.split('_')[0])-1]
    shin_frames.append(shin_excel)
    
```

```python
shin_frames
```

```python
dang_result = pd.concat(dang_frames)
shin_result = pd.concat(shin_frames)
```

```python
dang_result['접수번호'] = dang_result['접수번호'].apply(lambda x : ''+str(x))  ## lambda로 숫자를 문자로 변환
shin_result['접수번호'] = shin_result['접수번호'].apply(lambda x : ''+str(x))  ## lambda로 숫자를 문자로 변환
```

```python
dang_result.head()
```

```python
dang_result['접수번호_블록'] = dang_result['접수번호']+'_'+dang_result['블록']
shin_result['접수번호_블록'] = shin_result['접수번호']+'_'+shin_result['블록']
```

```python
# pd.merge(df_new, df_n, on='subject_id')
merge_data = pd.merge(shin_result ,dang_result, on = '접수번호_블록', how='inner')
```

```python
merge_data
```

```python
merge_data.columns = merge_data.columns.str.replace(' ', '_')
```

```python
merge_data['성명'] = merge_data['성명'].str[0:2]+"*"
merge_data['신청인_성명'] = merge_data['신청인_성명'].str[0:2]+"*"
merge_data['신청인_주민번호'] = merge_data['신청인_주민번호'].str[0:8]+"******"
merge_data['배우자_주민번호'] = merge_data['배우자_주민번호'].str[0:8]+"******"
merge_data['연락처_핸드폰1'] = ''
merge_data['연락처_핸드폰2'] = ''
merge_data['연락처_집전화'] = ''
merge_data['우편번호'] = merge_data['우편번호'].str[0:3]+"**"
```
